```python
#autoskalowanie danych
import numpy as np
def autoskalowanie(x):
    śr=x.mean()
    s=x.std()
    for i in x:
        a=(i-śr)/s
        print(a)
```

```python
dw = pd.read_excel(r'.\Moje-dane_MP.xlsx')
n=dw.iloc[2:,1:]
nowe=n.rename(columns={'Median\nweekly\nearnings':'Mediana zarobków kobiet','Median\nweekly\nearnings.1':'Mediana zarobków mężczyzn'})
nowe
```

```python
macierz=np.zeros((49,49))
nowe1=pd.DataFrame(nowe)
```

```python
#tworzenie macierzy odległości
#euklidesowa
sum=0
for i in range(len(nowe1)):
    for j in range(len(nowe1)):
        for k in range(0,6):
            sum += (nowe1.iloc[i,k]-nowe1.iloc[j,k])**2
        macierz[i][j]=np.sqrt(sum)
        sum=0
```

```python
euklides=pd.DataFrame(macierz)
euklides
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("dendrogramy metodą Warda")
d=shc.dendrogram(shc.linkage(nowe1, method='ward'))
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("dendrogramy metodą Warda")
d=shc.dendrogram(shc.linkage(euklides, method='ward'))
```

```python
plt.figure(figsize=(10,7))
plt.title("dendrogramy metodą najdalszego sąsiada")
d=shc.dendrogram(shc.linkage(nowe1, method='complete'))
```

```python
plt.figure(figsize=(15,5))
plt.title("dendrogramy metodą najbliższego sąsiada")
d=shc.dendrogram(shc.linkage(nowe1, method='single'))
```
